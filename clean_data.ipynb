```python
# Add dependencies
import glob
import pandas as pd
```

```python
# glob.glob('data*.csv') - returns List[str]
# pd.read_csv(f) - returns pd.DataFrame()
# for f in glob.glob() - returns a List[DataFrames]
# pd.concat() - returns one pd.DataFrame()
clean_data_ridership = pd.concat([pd.read_csv(f) for f in glob.glob('raw_data/rider_and_membership_data/*.csv')], ignore_index = True)
clean_data_trip = pd.concat([pd.read_csv(f) for f in glob.glob('raw_data/bike_trip_data/*.csv')], ignore_index = True)
```

```python
#dropping rows with na values
clean_data_trip = clean_data_trip.dropna(subset=["start station id",
                                                 "start station name",
                                                 "end station id",
                                                 "end station name"]
                                        )
```
